```python
import pandas as pd
import numpy as np
import re
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('preimputed_ptb_risk_file.csv')
```

```python
df.query(" (MT_BABIES_COUNT == 1) and (TERM != -1)")
```

```python
df.columns
```

```python
#Preprocessing ===============

##Target column has all zeros:preterm_lscs_emergency
df=df.drop(columns='preterm_lscs_emergency')
target_columns=['TERM','preterm_lscs','preterm_nvd']
df_upd=df.drop(columns=target_columns)
df_upd['Abortion']=df_upd['Abortion'].replace({"1L0":"1"})
df_upd['HB']=df_upd['HB'].replace({"10..2":"10.2",'11..9':'11.9','10 .1':'10.1'})
df_upd['HB']=df_upd['HB'].astype(float)
```

```python
missing_val_columns=['MT_WEIGHT (KG)', 'MT_HEIGHT (CM)','MT_AGE (YEARS)','PAROUS','HB','E.F.W', 'FHR', 'AFI', 'Grade', 'Cervix (cm)','Preeclampsia']
idvs = ['MT_WEIGHT (KG)', 'MT_AGE (YEARS)','MT_HEIGHT (CM)',
        'PAROUS', 'PRVS_LSCS', 'HYPOTHYROIDISM',
        'IVF/IUI', 'PPROM', 'OLIOGOHYDRAMNIOS', 'Abortion', 'HB',
       'Preeclampsia', 'E.F.W', 'FHR', 'AFI', 'Grade', 'Cervix (cm)',
       'Anterior Placenta', 'Posterior Placenta', 'Fundal Placenta',
       'Lateral Placenta', 'Right Lateral Posterior Placenta',
       'Fundo-Anterior Placenta', 'Fundo-Posterior Placenta', 'Separation',
       'Transverse Position', 'Cephalic Position', 'Breech Position','Planceta_PREVIA',
       'Hypertension']
print("Count of missing columns",len(missing_val_columns))
```

```python
df_upd['Preeclampsia'][df_upd['Preeclampsia']=='#VALUE!']=np.nan
```

```python
#Impute missing values using linear regression

df_impute=df_upd[idvs].copy()
# Step 2: Identify columns with missing values
missing_columns = missing_val_columns

for col in missing_columns:
        print(col)
        df_missing = df_impute[df_impute[col].isnull()]
        df_not_missing = df_impute[df_impute[col].notnull()]

        # Features (X) and target (y)
        X = df_not_missing.drop(columns=missing_columns)  # Exclude the missing column
        y = df_not_missing[col]

        # Step 4: Train a regression model
        model = LinearRegression()
        model.fit(X, y)

        # Step 5: Predict the missing values
        X_missing = df_missing.drop(columns=missing_columns)  # Exclude the missing column
        predicted_values = model.predict(X_missing)

        # Fill in the missing values
        df_impute.loc[df_impute[col].isnull(), col] = predicted_values

# Now df has the missing values replaced
```

```python
remaining_cols = list(set(df.columns.tolist()).difference(set(idvs)))
```

```python
#Concatenate the target variables
df_impute=pd.concat([df_impute,df[target_columns + remaining_cols]],axis=1)
df_upd=pd.concat([df_upd,df[target_columns]],axis=1)
```

```python
# Step 1: Calculate mean and standard deviation for non-imputed DataFrame
df_non_imputed_mean = df_upd[missing_columns].describe().loc[['mean', 'std']].T.reset_index()
df_non_imputed_mean.columns = ['column', 'mean_after', 'std_after']

# Step 2: Calculate mean and standard deviation for imputed DataFrame
df_imputed_mean = df_impute[missing_columns].describe().loc[['mean', 'std']].T.reset_index()
df_imputed_mean.columns = ['column', 'mean_before', 'std_before']

# Step 3: Calculate mode for both DataFrames
mode_non_imputed = df_upd[missing_columns].mode().iloc[0].reset_index()
mode_imputed = df_impute[missing_columns].mode().iloc[0].reset_index()

mode_non_imputed.columns = ['column', 'mode_after']
mode_imputed.columns = ['column', 'mode_before']


final_correlation_summary = pd.DataFrame()

for target in target_columns:
#     if target=="correlation_before_preterm_lscs_emergency":
        print(f"Processing target: {target}")

        # Step 1: Calculate correlation for the non-imputed DataFrame
        correlation_non_imputed = df_upd.corr(method='pearson')
        correlation_target_non_imputed = correlation_non_imputed[target].reset_index()
        col_name_before = 'correlation_before_' + target
        correlation_target_non_imputed.columns = ['column', col_name_before]

        # Step 2: Calculate correlation for the imputed DataFrame
        correlation_imputed = df_impute.corr(method='pearson')
        correlation_target_imputed = correlation_imputed[target].reset_index()
        col_name_after = 'correlation_after_' + target
        correlation_target_imputed.columns = ['column', col_name_after]

        # Step 3: Merge results
        correlation_summary = pd.merge(correlation_target_non_imputed, correlation_target_imputed, on='column', how='outer')

        # Filter to keep only relevant columns
        correlation_summary = correlation_summary[correlation_summary['column'].isin(missing_columns)]
        
        correlation_summary.rename(columns={'column':'column'+target},inplace=True)
        # Step 4: Concatenate results
        if final_correlation_summary.empty:
            final_correlation_summary = correlation_summary
        else:
            final_correlation_summary = pd.concat([final_correlation_summary, correlation_summary], axis=1)

subset_col=['columnTERM', 'correlation_before_TERM', 'correlation_after_TERM',
           'correlation_before_preterm_lscs','correlation_after_preterm_lscs',
           'correlation_before_preterm_nvd', 'correlation_after_preterm_nvd']
final_correlation_summary=final_correlation_summary[subset_col]
final_correlation_summary.rename(columns={'columnTERM':'column'},inplace=True)
# Step 4: Merge results
df_summary = pd.merge(df_non_imputed_mean, df_imputed_mean, on='column', how='outer')
df_summary = pd.merge(df_summary, mode_non_imputed, on='column', how='outer')
df_summary = pd.merge(df_summary, mode_imputed, on='column', how='outer')
df_summary = pd.merge(df_summary, final_correlation_summary, on='column', how='left')


columns=['column','mean_before','mean_after','std_before','std_after','mode_before','mode_after','correlation_before_TERM',
       'correlation_after_TERM', 'correlation_before_preterm_lscs',
       'correlation_after_preterm_lscs', 'correlation_before_preterm_nvd',
       'correlation_after_preterm_nvd']
df_summary=df_summary[columns]
df_summary = df_summary.round(2)
```

```python
df_summary.to_csv("summary.csv")
```
